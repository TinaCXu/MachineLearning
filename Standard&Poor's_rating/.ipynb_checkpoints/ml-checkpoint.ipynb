```python
import pandas as pd
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
import math
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn import SVM
# from xgboost.sklearn import XGBRegressor
# import xgboost as xgb
```

```python
def preprocess_data(data):
    data = data.drop('FX_RATE', axis=1)
    data = data.drop('ENTITY_NAME', axis=1)
    #data = data.drop('ACCOUNTING_STANDARD', axis=1)
    data = data.drop('CONSOLIDATION_TYPE', axis=1)
    data = data.drop('CURRENCY', axis=1)
    print(data)
    data = data.drop('ID', axis=1)
    data.ACCOUNTING_STANDARD = pd.factorize(data.ACCOUNTING_STANDARD)[0]
    data.RATING = pd.factorize(data.RATING)[0]
    print(data.RATING.value_counts())
    return data

def oversample(data):
    for value in data.RATING.unique():
        while len(data[data.RATING == value]) <= 154:
            new_sample = data[data.RATING == value].sample(n = 1)
            data = data.append(new_sample)
    return data
```

```python
#read data
f = open('data.csv', encoding='utf-8')
data = pd.read_csv(f)
data = preprocess_data(data)
#shuffle
data = data.sample(frac=1)
#divide into test and train
train_size = int(len(data) * 0.8)
train_data = data.iloc[:train_size, :]
test_data = data.iloc[train_size:, :]
#oversampling train data
train_data = oversample(train_data)
#over_sampled_test_data = oversample(test_data)
print (len(train_data), len(test_data))
```

```python
print (data.columns)
```

```python
#no oversample
#data = oversample(data) #comment this line if no oversample
y = data['RATING']
x = data.drop('RATING', axis=1)
#print(len(x), len(y))
print(y.value_counts())
x_train, x_test, y_train , y_test=model_selection.train_test_split(x,y,test_size=0.2)
print(x_test.shape)
print(y_test.shape)
print(x_train.shape)
print(y_train.shape)
print(y_train.value_counts(), y_test.value_counts())
```

```python
#only oversample train
y = train_data['RATING']
x = train_data.drop('RATING', axis=1)
x_train, _, y_train, _=model_selection.train_test_split(x,y,test_size=0.01)

y = test_data['RATING']
x = test_data.drop('RATING', axis=1)
_, x_test,_ , y_test=model_selection.train_test_split(x,y,test_size=0.99)

print(x_test.shape)
print(y_test.shape)
print(x_train.shape)
print(y_train.shape)
```

```python
#logistic
lr=LogisticRegression(max_iter=20000)
lr.fit(x_train,y_train)
y_pred = lr.predict(x_train)
y_test_pred = lr.predict(x_test)
accuracy = accuracy_score(y_train, y_pred)
print (accuracy)
accuracy = accuracy_score(y_test, y_test_pred)
print (accuracy)
```

```python
from sklearn.svm import SVC
clf = SVC(probability=True, kernel="rbf", C=1.0, gamma=0.01, max_iter=30000)
clf.fit(x_train, y_train)

y_pred = clf.predict(x_train)
accuracy = accuracy_score(y_train, y_pred)
print(accuracy)

y_pred = clf.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.preprocessing import OneHotEncoder, LabelBinarizer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_curve, auc, confusion_matrix
from sklearn.model_selection import GridSearchCV

models = []
models.append(('Logistic Regression', LogisticRegression()))
models.append(('GradientBoosting', GradientBoostingClassifier(n_estimators=21)))
models.append(('Naive Bayes', GaussianNB()))
models.append(('Decision Tree (NoParam)', DecisionTreeClassifier())) 
models.append(('Decision Tree (GridSearch)', GridSearchCV(DecisionTreeClassifier(), {'max_depth':[5, 10, 15, 20, 25, 32]}, cv=5)))
models.append(('RandomForestClassifier (GridSearch)', GridSearchCV(RandomForestClassifier(), {'max_depth':[5, 15], 'n_estimators':[20,22]})))
models.append(('RandomForestClassifier (2 Param)', RandomForestClassifier(n_estimators=22, criterion='entropy')))
models.append(('KNeighborsClassifier', KNeighborsClassifier(n_neighbors=5, metric='minkowski')))
# models.append(('Support Vector Regression', SVR(kernel='rbf')))
models.append(('LinearDiscriminantAnalysis', LinearDiscriminantAnalysis()))
models.append(('AdaBoostClassifier', AdaBoostClassifier(learning_rate=0.5)))
models.append(('BaggingClassifier', BaggingClassifier()))
models.append(('SVM', SVC(probability=True, kernel="rbf", C=1.0, gamma=0.01)))
model_name = []
acc_score = []
from sklearn.metrics import classification_report
# Modelleri test edelim
for name, model in models:
    model = model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    from sklearn import metrics
    # print("%s -> ACC: %%%.2f" % (name,metrics.accuracy_score(y_test, y_pred)*100))
    model_name.append(name)
    acc_score.append((metrics.accuracy_score(y_test, y_pred)*100))
    print('Model: ', name)
    print('Confusion Matrix: \n', metrics.confusion_matrix(y_test, y_pred))
    #Detaylı sınıflandırma raporu...
    report = classification_report(y_test, y_pred)
    print('\n ',report)
 
```
